# üß¨ Analyse simple de s√©quences d'ARNm en Python

## 1. Distribution des bases A, C, G, U

```
sequences = [
    "AUGCGAACGUUAGC",
    "AUGCGAACGUUAGC",
    "AUGCGAACGUUCGC",
    "AUGCGAACGUUUAC",
    "AUGCGAACGCUAGC"
]

from collections import Counter
import matplotlib.pyplot as plt

# Regrouper toutes les bases ensemble
toutes_les_bases = "".join(sequences)

# Compter les occurrences de chaque base
comptage_bases = Counter(toutes_les_bases)

# Affichage des r√©sultats
print("Occurrences des bases :")
for base in "ACGU":
    print(f"{base} : {comptage_bases[base]}")

# Graphique de distribution
bases = ["A", "C", "G", "U"]
valeurs = [comptage_bases[base] for base in bases]

plt.bar(bases, valeurs, color=["green", "blue", "orange", "red"])
plt.title("Distribution des bases A, C, G, U")
plt.ylabel("Occurrences")
plt.show()
```

## 2. Distribution des codons

```
# Table simplifi√©e de traduction codon -> acide amin√©
table_codons = {
    "AUG": "Met",
    "CGA": "Arg",
    "ACG": "Thr",
    "UUA": "Leu",
    "AGC": "Ser",
    "UAC": "Tyr",
    "UCG": "Ser",
    "UUU": "Phe"
}

# Extraire tous les codons
codons = []
for seq in sequences:
    for i in range(0, len(seq) - 2, 3):
        codon = seq[i:i+3]
        codons.append(codon)

# Compter les codons
comptage_codons = Counter(codons)

# Affichage des fr√©quences
print("\nOccurrences des codons :")
for codon, count in comptage_codons.items():
    print(f"{codon} : {count}")

# Graphique de distribution
labels, valeurs = zip(*comptage_codons.items())

plt.bar(labels, valeurs, color="purple")
plt.title("Distribution des codons")
plt.xlabel("Codon")
plt.ylabel("Occurrences")
plt.show()
```

## 3. Traduction des s√©quences en cha√Ænes d'acides amin√©s

```
print("\nTraduction des s√©quences :")
for seq in sequences:
    proteine = []
    for i in range(0, len(seq) - 2, 3):
        codon = seq[i:i+3]
        aa = table_codons.get(codon, "?")
        proteine.append(aa)
    print(f"S√©quence d'ARNm : {seq}")
    print(f"Prot√©ine correspondante : {'-'.join(proteine)}\n")
```
