# üåÄ Fractale de Mandelbrot (aper√ßu math√©matique)

Dans ce notebook, vous allez voir comment quelques lignes de Python permettent de g√©n√©rer une visualisation color√©e de la c√©l√®bre fractale de Mandelbrot. Pas besoin de tout comprendre le code pour l'instant ‚Äî amusez-vous √† ex√©cuter les cellules et observez le r√©sultat!

```
import numpy as np
import matplotlib.pyplot as plt

# D√©finir la r√©solution de l'image
largeur, hauteur = 800, 800

# D√©finir les limites du plan complexe
xmin, xmax = -2.0, 1.0
ymin, ymax = -1.5, 1.5

# Nombre d'it√©rations maximal
max_iter = 100

# Cr√©er la grille de points
x = np.linspace(xmin, xmax, largeur)
y = np.linspace(ymin, ymax, hauteur)
X, Y = np.meshgrid(x, y)
C = X + 1j * Y

# Initialiser le tableau Z et le tableau pour stocker le nombre d'it√©rations
Z = np.zeros_like(C, dtype=np.complex128)
iterations = np.zeros(C.shape, dtype=int)

# Calcul des it√©rations
for i in range(max_iter):
    mask = np.abs(Z) <= 2
    Z[mask] = Z[mask] ** 2 + C[mask]
    iterations[mask] = i

# Affichage
plt.figure(figsize=(8,8))
plt.imshow(iterations, cmap="hot", extent=(xmin, xmax, ymin, ymax))
plt.title("Fractale de Mandelbrot (Aper√ßu)")
plt.xlabel("Re")
plt.ylabel("Im")
plt.colorbar(label="It√©rations avant divergence")
plt.show()
```
